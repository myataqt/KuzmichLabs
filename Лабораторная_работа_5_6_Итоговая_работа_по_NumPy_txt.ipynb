**Задание 1.** *Вычислите значения функции* $$f(x)=\log_{x^2+1}\dfrac{e^{5x+1}+\sqrt{x^5}+1}{|3\cos(2\pi x)+2\sin(x+\pi/12)|}$$
при $x=1, 10, 100.$

```python
import math

def compute_function(x):
  return math.log((math.e**(5*x + 1) + math.sqrt(x**5) + 1) / abs(3*math.cos(2*math.pi*x) + 2*math.sin(x+math.pi/12)), x**2 + 1)

print(compute_function(1))
print(compute_function(10))
print(compute_function(100))
```

**Задание 2** *Вычислите значения функции при изменении аргумента в некоторых приделах с заданным шагом.*

1) $f(x)=x^2$ на отрезке $[-2; 2]$ с шагом $0,01,$ с шагом $0,1,$ с шагом $0,25.$

2) $f(x)=\cos(2\pi t)$ на отрезке $[-10; 10]$ с шагом $0,25,$ с шагом $1.$

3) $f(x)=2\cos(t − 2) + \sin(2 t − 4)$ на отрезке $t\in [{}−20\pi; 10\pi]$ с шагом $\pi$ и с шагом $\dfrac{\pi}{12}\,.$

4) $f(x)=\log_2(|x|)$ на числовом множестве $[-4; 0)\cup (0; 4]$ с шагом $0,1,$ с шагом $0,25.$

5) $f(x)=\sqrt[5]{x}$ на отрезке $[1; 32]$ с шагом $1$ и с шагом $0.25.$

```python
import numpy as np

def f1(x):
    return x**2

for step in [0.01, 0.1, 0.25]:
    x_values = np.arange(-2, 2+step, step)
    y_values = f1(x_values)
    print(f"For step {step}, f(x) values are {y_values}")

def f2(t):
    return np.cos(2*np.pi*t)

for step in [0.25, 1]:
    t_values = np.arange(-10, 10+step, step)
    y_values = f2(t_values)
    print(f"For step {step}, f(x) values are {y_values}")

def f3(t):
    return 2*np.cos(t - 2) + np.sin(2*t - 4)

for step in [np.pi, np.pi/12]:
    t_values = np.arange(-20*np.pi, 10*np.pi+step, step)
    y_values = f3(t_values)
    print(f"For step {step}, f(x) values are {y_values}")

def f4(x):
    return np.log2(np.abs(x))

for step in [0.1, 0.25]:
    x_values = np.concatenate([np.arange(-4, 0, step), np.arange(step, 4+step, step)])
    y_values = f4(x_values)
    print(f"For step {step}, f(x) values are {y_values}")

def f5(x):
    return np.power(x, 1/5)

for step in [1, 0.25]:
    x_values = np.arange(1, 32+step, step)
    y_values = f5(x_values)
    print(f"For step {step}, f(x) values are {y_values}")
```

**Задание 4.** *Решите системы линейных уравнений* $\,A^2X=B\,$ *и* $\,A^3A^{T}A^2X=B,\,$ *если*
$$
A=\left(\!
\begin{array}{cccc}
2 & 4 & 7 & 4 \\
4 & 1 & 6 & 2 \\
8 & 3 & 6 & 7 \\
6 & 3 & 5 & 7
\end{array}\!\right),
\qquad
B = \left(
\begin{array}{c}
2 \\
0 \\
3 \\
1
\end{array}\right).
$$

```python
A = np.array([[2, 4, 7, 4], [4, 1, 6, 2], [8, 3, 6, 7], [6, 3, 5, 7]])
B = np.array([2, 0, 3, 1])

A2 = np.dot(A, A)

X1 = np.linalg.solve(A2, B)

print("Solution to the system A^2X = B is", X1)

A3AT_A2 = np.dot(np.dot(A, A), np.dot(A, A.T))

X2 = np.linalg.solve(A3AT_A2, B)

print("Solution to the system A^3A^TA^2X = B is", X2)
```

**Задание 5.** *Дан массив из* $\ N\leq 30\ $ *случайных целых чисел. Вычислите:*

* сумму элементов массива;
* минимальный и максимальный элементы массива;
* среднее квадратичное чисел массива;
* элемент массива, наиболее удаленный от среднего арифметического чисел массива.

```python
import numpy as np

np.random.seed(0)
arr = np.random.randint(1, 100, 30)

sum_arr = np.sum(arr)
print("Sum of the array elements: ", sum_arr)

min_arr = np.min(arr)
max_arr = np.max(arr)
print("Minimum element of the array: ", min_arr)
print("Maximum element of the array: ", max_arr)

mean_square_arr = np.mean(arr**2)
print("Mean square of the array numbers: ", mean_square_arr)

mean_arr = np.mean(arr)

diff_arr = np.abs(arr - mean_arr)

farthest_element = arr[np.argmax(diff_arr)]
print("Element farthest from the arithmetic mean: ", farthest_element)
```
